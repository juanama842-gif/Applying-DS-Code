```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer

df_free = pd.read_csv("/content/drive/MyDrive/FD_50.csv")

df_semantic = df_free.copy()
df_semantic = df_semantic.dropna(subset=["free_text"]).reset_index(drop=True)

df_semantic.head()
```

```python
embed_model = SentenceTransformer(
    "sentence-transformers/all-MiniLM-L6-v2",
    cache_folder="/content/drive/MyDrive/hf_cache",
    device="cpu"
)

print("Embedding model loaded on CPU.")
```

```python
def get_embedding(text):
    return embed_model.encode(text, normalize_embeddings=True)
```

```python
# ===== Stance anchors (used to determine NPS) =====
stance_anchors = {
    "Promoter": [
        "I would definitely recommend this bank to others.",
        "I would highly recommend this bank.",
        "I am very likely to recommend this bank.",
        "I am very satisfied and this bank fits my needs very well."
    ],
    "Passive": [
        "I might recommend this bank, but I have some reservations.",
        "I am neutral about recommending this bank."
    ],
    "Detractor": [
        "I would not recommend this bank.",
        "I am unlikely to recommend this bank.",
        "I would avoid recommending this bank.",
        "This bank does not meet my needs."
    ]
}
```

```python
# Convert anchors to vectors
import numpy as np

def build_anchor_embeddings(anchor_dict):
    anchor_embeddings = {}

    for label, texts in anchor_dict.items():
        # Each anchor sentence → embedding
        emb_list = [get_embedding(t) for t in texts]

        # Take the average as the "centroid vector" of this category
        anchor_embeddings[label] = np.mean(emb_list, axis=0)

    return anchor_embeddings
```

```python
# Generate anchor vectors
stance_anchor_embeddings = build_anchor_embeddings(stance_anchors)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# 1. Calculate the similarity between each text and the three stance anchors
def compute_similarity(text, anchor_embeddings):
    text_emb = get_embedding(text).reshape(1, -1)

    scores = {}
    for label, anchor_emb in anchor_embeddings.items():
        anchor_emb = anchor_emb.reshape(1, -1)
        scores[label] = cosine_similarity(text_emb, anchor_emb)[0][0]

    return scores
# 2. Calculate similarity for each free_text
results = df_semantic["free_text"].apply(
    lambda x: compute_similarity(x, stance_anchor_embeddings)
)

sim_df = pd.DataFrame(list(results))

# 3. Merge similarity scores
df_semantic = pd.concat([df_semantic, sim_df], axis=1)
```

```python
# 4. Get the initial classification based on the highest similarity
df_semantic["predicted_stance"] = df_semantic[
    ["Promoter", "Passive", "Detractor"]
].idxmax(axis=1)
```

```python

# 5. Rule-based correction: correct clear negative and hesitant expressions
def correct_stance(row):
    text = str(row["free_text"]).lower()

    # Clearly not recommending → Detractor
    detractor_patterns = [
        "wouldn't recommend",
        "would not recommend",
        "not recommend",
        "unlikely to recommend",
        "i'm unlikely to recommend",
        "i am unlikely to recommend",
        "avoid recommending",
        "would advise against",
        "does not meet my needs",
        "doesn't meet my needs"
    ]

    # Hesitant, uncertain, or reserved expressions → Passive
    passive_patterns = [
        "hesitant",
        "not sure",
        "might recommend",
        "some reservations",
        "mixed feelings",
        "depends",
        "neutral",
        "need more information",
        "would need more information",
        "acceptable but",
        "okay but",
        "not outstanding"
    ]

    for p in detractor_patterns:
        if p in text:
            return "Detractor"

    for p in passive_patterns:
        if p in text:
            return "Passive"

    return row["predicted_stance"]


df_semantic["predicted_stance_corrected"] = df_semantic.apply(correct_stance, axis=1)
```

```python
df_semantic = df_semantic.loc[:, ~df_semantic.columns.duplicated()]
```

```python
df_semantic[[
    "free_text",
    "Promoter",
    "Passive",
    "Detractor",
    "predicted_stance",
    "predicted_stance_corrected"
]].head(10)
```

```python
# Calculate NPS
counts = df_semantic["predicted_stance_corrected"].value_counts(normalize=True)

promoter = counts.get("Promoter", 0)
passive = counts.get("Passive", 0)
detractor = counts.get("Detractor", 0)

nps = (promoter - detractor) * 100

print("Promoter %:", round(promoter * 100, 2))
print("Passive %:", round(passive * 100, 2))
print("Detractor %:", round(detractor * 100, 2))
print("NPS:", round(nps, 2))
```

```python
from transformers import pipeline

sentiment_pipeline = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english",
    device=-1   # CPU
)

print("Sentiment model loaded.")
```

```python
import pandas as pd
df_free = pd.read_csv("/content/drive/MyDrive/HSBC_720.csv")

df_semantic = df_free.copy()
df_semantic = df_semantic.dropna(subset=["free_text"]).reset_index(drop=True)

df_semantic.head()
```

```python
# Classify sentiment
df_semantic["sentiment"] = df_semantic["free_text"].apply(
    lambda x: sentiment_pipeline(str(x))[0]["label"]
)

df_semantic[["free_text", "sentiment"]].head()
```

```python
# Classify to 3 Types
def sentiment_to_nps(text, sentiment):
    text = str(text).lower()

    # def negative
    if sentiment == "NEGATIVE":
        return "Detractor"

    # def positive
    if sentiment == "POSITIVE":
        if (
            "definitely recommend" in text or
            "highly recommend" in text or
            "wouldn't hesitate" in text or
            "very likely" in text
        ):
            return "Promoter"
        else:
            return "Passive"

    return "Passive"
```

```python
# Generate stance
df_semantic["predicted_stance"] = df_semantic.apply(
    lambda row: sentiment_to_nps(row["free_text"], row["sentiment"]),
    axis=1
)
```

```python
# Calculate NPS
counts = df_semantic["predicted_stance"].value_counts(normalize=True)

promoter = counts.get("Promoter", 0)
passive = counts.get("Passive", 0)
detractor = counts.get("Detractor", 0)

nps = (promoter - detractor) * 100

print("Promoter %:", round(promoter * 100, 2))
print("Passive %:", round(passive * 100, 2))
print("Detractor %:", round(detractor * 100, 2))
print("NPS:", round(nps, 2))
```
